```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
X,Y = make_classification(n_classes=2,n_samples=400,n_clusters_per_class=1,random_state=3,n_features=2,n_informative=2,n_redundant=0)
```

```python
Y[Y == 0] = -1 #broadcasting
print(Y)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
class SVM:
    
    def __init__(self,C=1.0):
        self.C = C
        self.W = 0
        self.b = 0
        
    def hingeLoss(self,W,b,X,Y):
        loss = 0.0
        
        loss += 0.5*np.dot(W,W.T)
        
        m = X.shape[0]
        
        for i in range(m):
            ti = Y[i]*(np.dot(W,X[i].T)+b)
            loss += self.C * max(0,(1-ti))
            
        return loss[0][0]
    
    def fit(self,X,Y,batch_size = 100,learning_rate = 0.001,maxitr = 300):
        
        no_of_features = X.shape[1]
        no_of_samples = X.shape[0]
        
        n = learning_rate
        c = self.C
        
        #Initialize the model parameters
        W = np.zeros((1,no_of_features))
        bias = 0
        
        #Initial Loss
        print(self.hingeLoss(W,bias,X,Y))
        
        #Training 
        #Weight and Bias update rule
        losses = []
        
        for i in range(maxitr):
            
            #Training loop
            l = self.hingeLoss(W,bias,X,Y)
            losses.append(l)
            
            ids = np.arange(no_of_samples)
            np.random.shuffle(ids)
            
            #Batch gradient descent with random shuffling
            for batch_id in range(0,no_of_samples,batch_size):
                
                #Assuming 0 gradient for the batch
                gradw = 0
                gradb = 0
                
                #Iterate over all egs in mini batch
                for j in range(batch_id, batch_id+batch_size):
                    if j<no_of_samples:
                        i = ids[j]
                        ti = Y[i]*(np.dot(W,X[i].T)+bias)
                        
                        if ti>1:
                            gradw += 0
                            gradb += 0
                        else:
                            gradw += c*Y[i]*X[i]
                            gradb += c*Y[i]
                            
                 
                #Gradient for the batch is ready! Now Updating W,B
                W = W - n*W + n*gradw
                bias = bias + n*gradb
                
        self.W = W
        self.b = bias
        return W,bias,losses
    
                 
```

```python
mySvm = SVM(C = 1000)
W,b,losses = mySvm.fit(X,Y,maxitr=100)
print(losses[0])
print(losses[-1])
```

```python
plt.plot(losses)
plt.show()
```

```python
def hyperplane(w1,w2,b):
    
    plt.figure(figsize=(10,8))
    
    x1 = np.linspace(-2,4,10)
    x2 = -(w1*x1+b)/w2 #WT + B = 0  
    
    xp = -(w1*x1+b+1)/w2 #WT + B = -1 
    xn = -(w1*x1+b-1)/w2 #WT + B = +1 
    
    plt.plot(x1,x2, label = "Hyperplane WX+B=0")
    plt.plot(x1,xp, label = "+ve Hyperplane WX+B=1")
    plt.plot(x1,xn, label = "-ve Hyperplane WX+B=-1")
    
    plt.scatter(X[:,0],X[:,1],c=Y)
    plt.show()
    
```

```python
hyperplane(W[0,0],W[0,1],b)
```
